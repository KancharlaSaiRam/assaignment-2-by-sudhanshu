```python
# it takes two args function,iterable
# filter function calls the specified function which returns boolean for each item of the specified iterable(list,tuple)
```

```python
def is_prime(x):
    for i in range(2,x):
        if x%i == 0:
            return False
        else:
            return True            
```

```python
tuple(filter(is_prime,range(10)))
```

```python
fil = filter(is_prime,range(10))
print(list(fil))
```

```python
def my_filter(f,t):
    for i in t:
        a = is_prime(i)
        if a == True:
            yield i # generator 
```

```python
t=[12,3,5,34,123,345]
list(my_filter(is_prime,t))
```
